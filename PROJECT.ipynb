### 3.1 Tensor flow neural network

```python
import tensorflow as tf
```

```python
x = tf.placeholder(tf.float32, [None, k_train.shape[1]])
y = tf.placeholder(tf.float32, [None, m_train.shape[1]])

n_input= k_train.shape[1]
# #===added by yang====#
# hlayer_num = 10
# nerons_layer = {}
# for n in range(1,hlayer_num+1):
#     nerons_layer[n] = 10 
# #===added by yang====#
n_hidden_1 = 10
n_hidden_2 = 10
n_hidden_3 = 10
n_hidden_4 = 10
n_hidden_5 = 10
n_hidden_6 = 10
n_hidden_7 = 10
n_hidden_8 = 10
n_hidden_9 = 10
n_hidden_10 = 10
n_classes = m_train.shape[1]
total_len = 500
batch_size = 100

def multilayer_perceptron(x, weights, biases):
  
    layer_1 = tf.add(tf.matmul(x, weights['h1']), biases['b1'])
    layer_1 = tf.nn.relu(layer_1)

 
    layer_2 = tf.add(tf.matmul(layer_1, weights['h2']), biases['b2'])
    layer_2 = tf.nn.relu(layer_2)

#     # Hidden layer with RELU activation
    layer_3 = tf.add(tf.matmul(layer_2, weights['h3']), biases['b3'])
    layer_3 = tf.nn.relu(layer_3)

#     # Hidden layer with RELU activation
    layer_4 = tf.add(tf.matmul(layer_3, weights['h4']), biases['b4'])
    layer_4 = tf.nn.relu(layer_4)
    
    layer_5 = tf.add(tf.matmul(layer_4, weights['h5']), biases['b5'])
    layer_5 = tf.nn.relu(layer_5)
    
    layer_6 = tf.add(tf.matmul(layer_5, weights['h6']), biases['b6'])
    layer_6 = tf.nn.relu(layer_6)

    layer_7 = tf.add(tf.matmul(layer_6, weights['h7']), biases['b7'])
    layer_7 = tf.nn.relu(layer_7)

    layer_8 = tf.add(tf.matmul(layer_7, weights['h8']), biases['b8'])
    layer_8 = tf.nn.relu(layer_8)
    
    layer_9 = tf.add(tf.matmul(layer_8, weights['h9']), biases['b9'])
    layer_9 = tf.nn.relu(layer_9)

    layer_10 = tf.add(tf.matmul(layer_9, weights['h10']), biases['b10'])
    layer_10 = tf.nn.relu(layer_10)


    out_layer = tf.matmul(layer_10, weights['out']) + biases['out']
    return out_layer
```

```python
weights = {
    'h1': tf.Variable(tf.random_normal([n_input, n_hidden_1], 0, 0.1)),
    'h2': tf.Variable(tf.random_normal([n_hidden_1, n_hidden_2], 0, 0.1)),
    'h3': tf.Variable(tf.random_normal([n_hidden_2, n_hidden_3], 0, 0.1)),
    'h4': tf.Variable(tf.random_normal([n_hidden_3, n_hidden_4], 0, 0.1)),
    'h5': tf.Variable(tf.random_normal([n_hidden_4, n_hidden_5], 0, 0.1)),
    'h6': tf.Variable(tf.random_normal([n_hidden_5, n_hidden_6], 0, 0.1)),
    'h7': tf.Variable(tf.random_normal([n_hidden_6, n_hidden_7], 0, 0.1)),
    'h8': tf.Variable(tf.random_normal([n_hidden_7, n_hidden_8], 0, 0.1)),
    'h9': tf.Variable(tf.random_normal([n_hidden_8, n_hidden_9], 0, 0.1)),
    'h10': tf.Variable(tf.random_normal([n_hidden_9, n_hidden_10], 0, 0.1)),
    'out': tf.Variable(tf.random_normal([n_hidden_10, n_classes], 0, 0.1))
}
biases = {
    'b1': tf.Variable(tf.random_normal([n_hidden_1], 0, 0.1)),
    'b2': tf.Variable(tf.random_normal([n_hidden_2], 0, 0.1)),
    'b3': tf.Variable(tf.random_normal([n_hidden_3], 0, 0.1)),
    'b4': tf.Variable(tf.random_normal([n_hidden_4], 0, 0.1)),
    'b5': tf.Variable(tf.random_normal([n_hidden_5], 0, 0.1)),
    'b6': tf.Variable(tf.random_normal([n_hidden_6], 0, 0.1)),
    'b7': tf.Variable(tf.random_normal([n_hidden_7], 0, 0.1)),
    'b8': tf.Variable(tf.random_normal([n_hidden_8], 0, 0.1)),
    'b9': tf.Variable(tf.random_normal([n_hidden_9], 0, 0.1)),
    'b10': tf.Variable(tf.random_normal([n_hidden_10], 0, 0.1)),
    'out': tf.Variable(tf.random_normal([n_classes], 0, 0.1))
}
```

```python
pred = multilayer_perceptron(x, weights, biases)
```

```python
cost = tf.reduce_mean(tf.square(pred-y))
##cost = tf.reduce_mean(tf.nn.softmax_cross_entropy_with_logits(logits=pred,labels=y))
optimizer = tf.train.AdamOptimizer(learning_rate=0.01).minimize(cost)
```

```python
with tf.Session() as sess:
    sess.run(tf.initialize_all_variables())
    for epoch in range(1000):
        avg_cost = 0.
        total_batch = int(total_len/batch_size)
        for i in range(total_batch-1):
            batch_x = k_train[i*batch_size:(i+1)*batch_size]
            batch_y = m_train[i*batch_size:(i+1)*batch_size]
            _, c, p = sess.run([optimizer, cost, pred], feed_dict={x: batch_x, y: batch_y})
            avg_cost += c / total_batch
        label_value = batch_y
        estimate = p
        err = label_value-estimate
        if epoch % 10 == 0:
            print ("Epoch:", '%04d' % (epoch+1), "cost=", \
                "{:.9f}".format(avg_cost))
            print ("[*]----------------------------")
    print ("done")
    correct_prediction = tf.equal(tf.argmax(pred, 1), tf.argmax(y, 1))
    accuracy = tf.reduce_mean(tf.cast(correct_prediction, "float"))
    print ("Accuracy:", accuracy.eval({x: k_test, y: m_test}))
```
